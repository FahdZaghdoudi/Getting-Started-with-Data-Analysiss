## Understanding student's online behavior and their relations inside real class gradings

### About our Dataset
This Dataset was compiled after 4 months of an Algorithm Introductory Class at a Brazilian University.

A traditional grading system was adopted as evaluation of the students performance, and, at the same time, an online environment let students share posts, answers and classify productions with emojis-based reactions.

The Class was project-based and the evaluation of the skills followed the so-called "21th Century Skills", in a scale from 0 to 10 each Skill:

   * Critical Thinking and Problem Solving Skills - named as SK1;
   * Criativity and Inovation Skills - named as SK2;
   * Constant and Self Learning Skills - named as SK3;
   * Collaboration and Self-Direction Skills - named as SK4;
   * Social and Cultural Responsability - named as SK5.

At the online learning environment, the students could post and classify colleagues postings with some reactions.

The Dataset contains, one unique student per row, the amount of reactions of each type received by the student by other student as well as his gradings in each Skills evaluated by the professor. 

## Columns
* Row Count
* total_posts : Amount of Posts created by the student
* helpful_post : Amount of Reactions "helpful" received by the student
* nice_code_post : Amount of Reactions "nice code" received by the student
* collaborative_post : Amount of Reactions "collaborative" received by the student
* confused_post : Amount of Reactions "confused" received by the student
* creative_post : Amount of Reactions "creative" received by the student
* bad_post : Amount of Reactions "bad" received by the student
* amazing_post : Amount of Reactions "amazing" received by the student
* timeonline : Amount of time (in seconds) spent by the student at the environment
* sk1_classroom : Final grade of the Skill 1 for the student
* sk2_classroom : Final grade of the Skill 2 for the student
* sk5_classroom : Final grade of the Skill 5 for the student
* sk3_classroom : Final grade of the Skll 3 for the student
* sk4_classroom : Final grade of the Skill 4 for the student
* Approved : Was the student approved at the Class? YES / NO

##  Correlation Matrix

we will first divide the dataset features into 2 subsets for a better manipulation and analysis of the data.The first subset is composed of features related to posts in the online environment and the second one is composed of the evaluation in each skill.

## Posts features subset analysis

```python
ps=df.iloc[:,0:8] #post subset 
ps_cor = ps.corr()
ps_cor
```

```python
#we can use also heatmap for a better understanding
ps_cor.style.background_gradient(cmap='coolwarm')
```

```python
#importing the principal component analysis library
from sklearn.decomposition import PCA 
 
#instantiation of a pca object from the library
acp = PCA(svd_solver='full') 
```

```python
#calculating
coord = acp.fit_transform(ps) 
#nnumber of componenets calculated
print(acp.n_components_)
```

```python
#calculating the eigenvalues (valeurs propres)
n=ps.shape[0]
p=ps.shape[1]
eigval = (n-1)/n*acp.explained_variance_ 
print(eigval) 
```

```python
#scree plot(plotting the eigen values)
import numpy
plt.plot(numpy.arange(1,p+1),eigval)
plt.title("Scree plot")
plt.ylabel("Eigen values")
plt.xlabel("Factor number")
plt.show() 
```

```python
#calculating the square root of eigenvalues
sqrt_eigval = numpy.sqrt(eigval) 
```

```python
#plotting individuals in the first plan
fig, axes = plt.subplots(figsize=(12,12))
axes.set_xlim(-6,6)
axes.set_ylim(-6,6)
for i in range(n):
    plt.annotate(ps.index[i],(coord[i,0],coord[i,1])) 

plt.plot([-6,6],[0,0],color='silver',linestyle='-',linewidth=1)
plt.plot([0,0],[-6,6],color='silver',linestyle='-',linewidth=1) 
 
#plotting
plt.show()
```

```python
#variable correlation with the axis
corvar = numpy.zeros((p,p)) 
 
for k in range(p):
    corvar[:,k] = acp.components_[k,:] * sqrt_eigval[k]

#correlation matrix of variables with factors
print(corvar)
```

```python
# there are two dominant axis
print(pd.DataFrame({'id':ps.columns,'COR_1':corvar[:,0],'COR_2':corvar[:,1]}))
```

```python
#correlation circle
fig, axes = plt.subplots(figsize=(8,8))
axes.set_xlim(-1,1)
axes.set_ylim(-1,1) 
 
#display of labels (variable names)
for j in range(p):
    plt.annotate(ps.columns[j],(corvar[j,0],corvar[j,1]))
#add axes
plt.plot([-1,1],[0,0],color='silver',linestyle='-',linewidth=1)
plt.plot([0,0],[-1,1],color='silver',linestyle='-',linewidth=1)
#add a circle
cercle = plt.Circle((0,0),1,color='blue',fill=False)
axes.add_artist(cercle) 
 
#display
plt.show()
```

## Data Spliting
We will split our data to :
* Training set (80%)
* Test set (20%)

```python
#first, we import train_test_split from sklearn
from sklearn.model_selection import train_test_split
#features
X = df.drop('Approved', axis=1)
#target
Y = df['Approved']
```

```python
X.head()
```

```python
Y.head()
```

```python
#now we define our variables x_train, x_test, y_train, y_test 
#the test size is taken as 20%
x_train,x_test,y_train,y_test = train_test_split(X, Y, test_size=0.2, random_state=77)
```

## Model Evaluation

<b>Random forest</b><br>
Random forests or random decision forests are an ensemble learning method for classification, regression and other tasks that operate by constructing a multitude of decision trees at training time and outputting the class that is the mode of the classes or mean prediction of the individual trees.

<img src='https://miro.medium.com/max/1184/1*i0o8mjFfCn-uD79-F1Cqkw.png'>

```python
#First we will try with 50 decision tree, and a Max depth of 5
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=50,max_depth=5)
```

```python
#Here we train our model
model.fit(x_train, y_train)
#And finally, we use the model , and store them on predict variable
predict=model.predict(x_test)
```

```python
#Now we test the accuracy of our model
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predict))
```
